````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import time
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Sending an HTTP GET request to the URL
page=requests.get(url)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
print(soup.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
# Find the elements that contain movie URLs
movie_elements = soup.find_all('a', class_='title-list-grid__item--link')

# Extract the URLs from the elements
movie_urls = [element['href'] for element in movie_elements]

# Print the movie URLs
print(movie_urls)
````

````python
# getting the complete url for movies
movie_urls_c = []
for u in movie_urls:

  movie_urls_c.append('https://www.justwatch.com' + u)

# Print the movie URLs
print(movie_urls_c)
````

## **Scrapping Movie Title**

````python
# Write Your Code here
movies = soup.find_all('img', class_='picture-comp__img')
movie_titles = [element['alt'] for element in movies]
print(movie_titles)
````

## **Scrapping release Year**

````python
your_movie_url = movie_urls_c
release_years = []

for url in your_movie_url:
    # Send a GET request to the movie URL
    response = requests.get(url)

# Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the movie page
        movie_soup = BeautifulSoup(response.text, 'html.parser')

        # Find the element that contains the movie year
        year_element = movie_soup.find('div', class_='title-block')

        # Extract the movie year
        movie_year = year_element.find_next('span', class_='text-muted').text.strip() if year_element else "Year not found"

        numeric_year = re.search(r'\d{4}', movie_year)
        release_years.append(numeric_year.group() if numeric_year else "Year not found")

    else:
        print(f"Failed to retrieve the movie page at {url}. Status code: {response.status_code}")

print(f"Movie Year: {release_years}")
````

## **Scrapping Genres**

````python
# Write Your Code here
your_movie_url = movie_urls_c
movie_genre = []

for url in your_movie_url:
    # Send a GET request to the movie URL
    response = requests.get(url)

    if response.status_code == 200:
      movie_soup = BeautifulSoup(response.text, 'html.parser')
      genre_element = movie_soup.find('h3', class_='detail-infos__subheading', string='Genres')
      genre_value = genre_element.find_next('div', class_='detail-infos__value').text.strip()

      movie_genre.append(genre_value)
    else:
        print(f"Failed to retrieve the genre at {url}. Status code: {response.status_code}")

print(f"Movie Genre: {movie_genre}")

````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
your_movie_url = movie_urls_c
imdb_rating = []

for url in your_movie_url:
  #print(url)
  # Send a GET request to the movie URL
  response = requests.get(url)
  if response.status_code == 200:
    movie_soup = BeautifulSoup(response.text, 'html.parser')
    imdb_rating_element = movie_soup.find('img', alt='IMDB')
    #print(imdb_rating_element)
    if imdb_rating_element:
      imdb = imdb_rating_element.find_next('span').text.strip()
    else:
      imdb = None
    #print(imdb)
    imdb_rating.append(imdb)
  else:
      print(f"Failed to retrieve the genre at {url}. Status code: {response.status_code}")

imdb_ratings = [float(rating.split()[0]) if rating else None for rating in imdb_rating]
print("IMDb Ratings as Numeric Values:", imdb_ratings)
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
# Write Your Code here
run_time_values = []
for url in your_movie_url:
  #print(url)
  # Send a GET request to the movie URL
  response = requests.get(url)
  if response.status_code == 200:
    movie_soup = BeautifulSoup(response.text, 'html.parser')
    runtime_element = movie_soup.find('h3', class_='detail-infos__subheading', string='Runtime')
    if runtime_element:
      runtime_element_value = runtime_element.find_next('div', class_='detail-infos__value').text.strip()
      #print(runtime_element_value)
    else:
      runtime_element_value = None
      #print(runtime_element_value)
    run_time_values.append(runtime_element_value)
  else:
      print(f"Failed to retrieve the runtime value at {url}. Status code: {response.status_code}")

print(f"Run time: {run_time_values}")
````

## **Scrapping Age Rating**

````python
# Write Your Code here
your_movie_url = movie_urls_c
age_ratings = []

for url in your_movie_url:
  #print(url)
    # Send a GET request to the movie URL
  response = requests.get(url)

  if response.status_code == 200:
    movie_soup = BeautifulSoup(response.text, 'html.parser')
    age_rating_element = movie_soup.find('h3', class_='detail-infos__subheading', string='Age rating')
    #print(age_rating_element)
    if age_rating_element:
      age_rating_value = age_rating_element.find_next('div', class_='detail-infos__value').get_text(strip=True)
      #print(age_rating_value)
    else:
      age_rating_value = None
    age_ratings.append(age_rating_value)
  else:
    print(f"Failed to retrieve the age_rating at {url}. Status code: {response.status_code}")
print(f"Age rating: {age_ratings}")
````

## **Fetching Production Countries Details**

````python
# Write Your Code here
your_movie_url = movie_urls_c
production_countries = []

for url in your_movie_url:
  #print(url)
  # Send a GET request to the movie URL
  response = requests.get(url)
  movie_soup = BeautifulSoup(response.text, 'html.parser')
  if response.status_code == 200:
    # Find all div elements with class "detail-infos"
    detail_infos_divs = movie_soup.find_all('div', class_='detail-infos')
    country_value = None
    # Iterate through each div and extract country value
    for detail_infos_div in detail_infos_divs:
      h3_element = detail_infos_div.find('h3', class_='detail-infos__subheading', style="max-width: fit-content")
      if h3_element and "Production country" in h3_element.text:
        country_value = detail_infos_div.find('div', class_='detail-infos__value').text.strip()
      else:
        country_value is None
    production_countries.append(country_value)
  else:
      print(f"Failed to retrieve the production country at {url}. Status code: {response.status_code}")

print(f"Production Country: {production_countries}")
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
your_movie_url = movie_urls_c
streaming_platform_names = []

for url in your_movie_url:
  #print(url)
    # Send a GET request to the movie URL
  response = requests.get(url)

  if response.status_code == 200:
    movie_soup = BeautifulSoup(response.text, 'html.parser')
    platform_name_element = movie_soup.find_all('img', class_='offer__icon')
    #print(age_rating_element)
    if platform_name_element:
      platform_name_value = [element['alt'] for element in platform_name_element]
      #print(platform_name_value)
    else:
      platform_name_value = None
    streaming_platform_names.append(platform_name_value)
  else:
    print(f"Failed to retrieve the streaming_platform_names  at {url}. Status code: {response.status_code}")
print(f"streaming platform : {streaming_platform_names}")
````

## **Now Creating Movies DataFrame**

````python
# Write Your Code here
import pandas as pd

data = {
    "movie_urls": [movie_urls_c],
    "movie_title": [movie_titles],
    "release_year": [release_years],
    "genre": [movie_genre],
    "IMDb_rating": [imdb_ratings],
    "production_country" :[production_countries],
    "streaming_services": [streaming_platform_names],
    "Runtime": [run_time_values],
    "age_rating": [age_ratings],
}
df = pd.DataFrame(data)
df_movie = df.apply(lambda col: col.explode())

# Reset the index
df_movie.reset_index(drop=True, inplace=True)

# Display the DataFrame
df_movie.head()
df_movie.info()
````

## **Task 2 :- Data Filtering & Analysis**

````python
# Write Your Code here
#- Only include movies and TV shows released in the last 2 years (from the current date).
# Only include movies and TV shows with an IMDb rating of 7 or higher.

from datetime import datetime, timedelta
df_movie['release_year'] = pd.to_datetime(df_movie['release_year'], format='%Y').dt.year
df_tv['release_year'] = pd.to_datetime(df_tv['release_year'], format ='%Y').dt.year
````

````python
current_year = datetime.now().year

df_movie_filter = df_movie[(df_movie['release_year'] >= current_year-2) & (df_movie['IMDb_rating'] >= 7)]
df_tv_filter = df_tv[(df_tv['release_year'] >= current_year-2) & (df_tv['IMDb_rating'] >= 7)]

print(df_movie_filter.shape, df_tv_filter.shape)
````

## **Calculating Mean IMDB Ratings for both Movies and Tv Shows**

````python
# Write Your Code here

mean_imdb_rating_movie = round(df_movie_filter['IMDb_rating'].mean(),2)
mean_imdb_rating_tv = round(df_tv_filter['IMDb_rating'].mean(),2)

print(mean_imdb_rating_movie)
print(mean_imdb_rating_tv)
````

## **Analyzing Top Genres**

````python
# Write Your Code here
# Concatenate the DataFrames
df_filter = pd.concat([df_movie_filter,df_tv_filter], ignore_index=True)

genre_counts = df_filter['genre'].value_counts()
# Count the occurrences of each genre
top5_genres = genre_counts.head(5)

print("Top 5 genres with the highest number of available movies and TV shows:")
print(top5_genres)
````
